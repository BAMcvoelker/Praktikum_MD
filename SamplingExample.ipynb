```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the number of samples
n_samples = 9

# Define the constraints for each dimension
constraints = [
    (350, 430),  # Dimension 1
    (0.45, 0.65),  # Dimension 2
    (0, 1)  # Dimension 3
]

# Sample the first dimension using LHS
first_dim = uniform.rvs(size=n_samples, loc=constraints[0][0], scale=constraints[0][1]-constraints[0][0])

# Sample the second dimension using LHS
second_dim = uniform.rvs(size=n_samples, loc=constraints[1][0], scale=constraints[1][1]-constraints[1][0])

# Sample the third dimension using random choice
categories = [0, 1]  # possible categories
third_dim = [random.choice(categories) for _ in range(n_samples)]

# Combine all the samples into one matrix
samples = np.vstack((first_dim, second_dim, third_dim)).T

# Use different markers for different categories
markers = ["o", "s"]
colors = ['blue','green']

for i in range(2):
    plt.scatter(samples[samples[:, 2] == i, 0], samples[samples[:, 2] == i, 1], 
                c=colors[i], marker=markers[i], label=categories[i], s=150)

# Add axis labels
plt.xlabel('Cement Content')
plt.ylabel('W/C-ratio')

# Add a title
plt.title('Latin Hypercube Sampling')

# add a legend
plt.legend(["CEMI", "CEMIII"])

# set grid
plt.grid(True, linestyle=':')

# set resolution
plt.gcf().set_size_inches(8, 6)

# display the plot
plt.show()
```
